## Main imports 

```python
import math
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data.sampler import SubsetRandomSampler
from torchvision import transforms, datasets, models
import matplotlib.pyplot as plt
from livelossplot import PlotLosses

device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
```

## Load images 

```python
input_shape = 100
rotation_degrees = 30
scale = 32
mean = (0.5,0.5,0.5)
std = (0.5,0.5,0.5)

# for training images apply some random transforms - rotaion, crop and flip, resize the image and normalize its channels
## the randomization will help in expanding the training data and improving generalization
train_images = datasets.ImageFolder('images', transform=transforms.Compose([
    transforms.RandomRotation(rotation_degrees),
    transforms.RandomResizedCrop(input_shape),
    transforms.RandomHorizontalFlip(),
    transforms.Resize(scale),
    transforms.ToTensor(),
    transforms.Normalize(mean, std)
]))

# for validation images apply the same transformations as the training images
valid_images = datasets.ImageFolder('images', transform=transforms.Compose([
    transforms.RandomRotation(rotation_degrees),
    transforms.RandomResizedCrop(input_shape),
    transforms.RandomHorizontalFlip(),
    transforms.Resize(scale),
    transforms.ToTensor(),
    transforms.Normalize(mean, std)
]))

# for testing images only resize the image and normalize its channels
test_images = datasets.ImageFolder('images', transform=transforms.Compose([
    transforms.Resize(scale),
    transforms.ToTensor(),
    transforms.Normalize(mean, std)
]))
```

## Create training, validation, testing subsets from the images dataset

```python
num_images = len(train_images)

indices = list(range(num_images))

# use 60% of the images dataset for training, 20% for testing and 20% validation

# define the spliting ranges
split60, split20 = int(np.floor(num_images * 0.6)), int(np.floor(num_images * 0.2))

# include randomness to the datasets
np.random.shuffle(indices)

# indices for each sub-dataset
train_indices = indices[:split60]
valid_indices = indices[split60:split60+split20]
test_indices = indices[split60+split20:]

# training, validation and testing datasets
train_samples = torch.utils.data.Subset(train_images, train_indices)
valid_samples = torch.utils.data.Subset(valid_images, valid_indices)
test_samples = torch.utils.data.Subset(test_images, test_indices)
```

## Load training, validation, testing data

```python
train_loader = torch.utils.data.DataLoader(train_samples, shuffle=True, batch_size=64, drop_last=True)

valid_loader = torch.utils.data.DataLoader(valid_samples, shuffle=True, batch_size=64, drop_last=True)

test_loader = torch.utils.data.DataLoader(test_samples, shuffle=False, batch_size=64, drop_last=True)
```

## Set an iterator for the datasets

```python
# helper function to make getting another batch of data easier
def cycle(iterable):
    while True:
        for x in iterable:
            yield x
       
class_names = train_images.classes
num_classes = len(class_names)

train_iterator = iter(cycle(train_loader))
valid_iterator = iter(cycle(valid_loader))
test_iterator = iter(cycle(test_loader))

print(f'> Size of training dataset {len(train_loader.dataset)}')
print(f'> Size of validation dataset {len(valid_loader.dataset)}')
print(f'> Size of testing dataset {len(test_loader.dataset)}')
```

## Convolutional Neural Network

```python
# Predefined CNN in pytorch
# N = models.resnet18()
# num_ftrs = N.fc.in_features
# N.fc = nn.Sequential(nn.Linear(num_ftrs, 512),
#                                  nn.ReLU(),
#                                  nn.Dropout(0.2),
#                                  nn.Linear(512, num_classes),
#                                  nn.LogSoftmax(dim=1))
# N.to(device)

# Basic CNN
class ConvolutionalNetwork(nn.Module):
    def __init__(self):
        super(ConvolutionalNetwork, self).__init__()
        layers = nn.ModuleList()
        layers.append(nn.Conv2d(3, 64, kernel_size=4, stride=2, padding=1, bias=False))
        layers.append(nn.BatchNorm2d(64))
        layers.append(nn.ReLU())
        layers.append(nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1, bias=False))
        layers.append(nn.BatchNorm2d(128))
        layers.append(nn.ReLU())
        layers.append(nn.Conv2d(128, 6, kernel_size=4, stride=2, padding=1, bias=False))
        layers.append(nn.AvgPool2d((4,4)))
        self.layers = layers

    def forward(self, x):
        for m in self.layers:
            x = m(x)
        return x

N = ConvolutionalNetwork().to(device)
```

## Set optimizer and learning rate

```python
lr = 1e-3 # set the learning rate

# initialise the optimiser
optimiser = torch.optim.Adam(N.parameters(), lr=lr)
```

## Model Training 

```python
best_model_wts = copy.deepcopy(N.state_dict())
best_acc = 0.0
epoch = 0
   
# use liveplot package for accuracy and loss plots
liveplot = PlotLosses()

while (epoch < 100):
    
    # arrays for metrics
    logs = {}
    train_loss_arr = np.zeros(0)
    train_acc_arr = np.zeros(0)
    valid_loss_arr = np.zeros(0)
    valid_acc_arr = np.zeros(0)
    

    # iterate over some of the train dateset
    for i in range(1000):
        x, t = next(train_iterator)
        x, t = x.to(device), t.to(device)
        
        optimiser.zero_grad()
        p = N(x).view(x.size(0), len(class_names))
        pred = p.argmax(dim=1, keepdim=True)
        loss = torch.nn.functional.cross_entropy(p, t)
        loss.backward()
        optimiser.step()

        train_loss_arr = np.append(train_loss_arr, loss.cpu().data)
        train_acc_arr = np.append(train_acc_arr, pred.data.eq(t.view_as(pred)).float().mean().item())

    # iterate entire valid dataset
    for i in range(100):
        x, t = next(valid_iterator)
        x,t = x.to(device), t.to(device)

        p = N(x).view(x.size(0), len(class_names))
        loss = torch.nn.functional.cross_entropy(p, t)
        pred = p.argmax(dim=1, keepdim=True)

        valid_loss_arr = np.append(valid_loss_arr, loss.cpu().data)
        valid_acc_arr = np.append(valid_acc_arr, pred.data.eq(t.view_as(pred)).float().mean().item())

    epoch_valid_acc = valid_acc_arr.mean()
        
    # keep track of the model weights for best validation accuracy
    if epoch_valid_acc > best_acc:
        best_acc = epoch_valid_acc
        best_model_wts = copy.deepcopy(N.state_dict())
        
    # update the accuracy and loss plots
    liveplot.update({
        'accuracy': train_acc_arr.mean(),
        'val_accuracy': epoch_valid_acc,
        'loss': train_loss_arr.mean(),
        'val_loss': valid_loss_arr.mean()
    })
    liveplot.draw()
    
    epoch = epoch + 1
    
torch.save(best_model_wts, 'land_classifier_model.pkl')
```

## Model Testing

```python
def plot_image(i, predictions_array, true_label, img):
    predictions_array, true_label, img = predictions_array[i], true_label[i], img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])

    plt.imshow(img, cmap=plt.cm.binary)

    predicted_label = np.argmax(predictions_array)
    color = '#335599' if predicted_label == true_label else '#ee4433'
    
    plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                  100*np.max(predictions_array),
                                  class_names[true_label]),
                                  color=color)

def plot_value_array(i, predictions_array, true_label):
    predictions_array, true_label = predictions_array[i], true_label[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    thisplot = plt.bar(range(len(class_names)), predictions_array, color="#777777")
    plt.ylim([0, 1]) 
    predicted_label = np.argmax(predictions_array)

    thisplot[predicted_label].set_color('#ee4433')
    thisplot[true_label].set_color('#335599')

test_images, test_labels = next(test_iterator)
test_images, test_labels = test_images.to(device), test_labels.to(device)
test_preds = torch.softmax(N(test_images).view(test_images.size(0), len(class_names)), dim=1).data.squeeze().cpu().numpy()

num_rows = 4
num_cols = 4
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
    plt.subplot(num_rows, 2*num_cols, 2*i+1)
    plot_image(i, test_preds, test_labels.cpu(), test_images.cpu().squeeze().permute(1,3,2,0).contiguous().permute(3,2,1,0))
    plt.subplot(num_rows, 2*num_cols, 2*i+2)
    plot_value_array(i, test_preds, test_labels)
```
